```python
import pandas as pd
import os
```

```python
def create_csv_files(path):
    csv_file_dict = {} 
    for each_csv in os.listdir(os.path.join('simplicity', 'data', '/')):  
        if each_csv not in os.listdir(path):
            csv_df_in = pd.read_csv(os.path.join(os.path.join('simplicity', 'data',
                                              each_csv)))
            csv_df_out = pd.DataFrame(columns = list(csv_df_in.columns))
            csv_df_out.to_csv(os.path.join(path, 
                                           each_csv),
                             index = None)
            csv_file_dict[each_csv] = list(csv_df_in.columns)
    
    return csv_file_dict
```

```python
create_csv_files(os.path.join('osemosys_global_model','data','/'))
```
